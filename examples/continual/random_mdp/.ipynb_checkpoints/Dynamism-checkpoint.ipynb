```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
```

```python
data = pickle.load(open('dynamism_2020_02_06_22_51_02.pkl', 'rb'))
```

```python
val_returns = data['val_rewards'].sum(axis=1)
times = data['times']
```

```python
plt.plot(np.array(data['variant']['betas']), val_returns / val_returns.max())
plt.title('Dynamism Performance in Random MDP (S=100, A=5)')
plt.xlabel('Beta (Degree of Dynamism)')
plt.ylabel('Validation Performance')
plt.show()
```

```python
plt.plot(np.array(data['variant']['betas']), times[0])#.mean(axis=0))
plt.title('Dynamism Performance in Random MDP (S=100, A=5)')
plt.xlabel('Beta (Degree of Dynamism)')
plt.ylabel('Q-Learning Convergence Time')
plt.show()
```

```python
times2 = times.copy()
times2[times2 == -1] = times2.max()
```
