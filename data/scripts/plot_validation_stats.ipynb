```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from os.path import join
import pickle
import pprint
```

```python
stats_dir = '/home/suvansh/levine/continualrl/data/scripts/validation_stats/'
exp_name = '09-08-tool-dqn-env-shaping-frequency-decrease'
stats_pkl_file = 'stats_2019_09_09_05_39_49.pkl'

stats_path = join(stats_dir, exp_name, stats_pkl_file)
stats = pickle.load(open(stats_path, 'rb'))
```

```python
# leave lists blank to choose all
desired_metrics = ['proportion_solved']
desired_variants = [198929321776461040, -161893430528816860, 9101241210143366543, 6904784455701835292, 8484378055337889386]
```

```python
# looping over variants (hyperparam settings)
xs = {}
ys = {}
for variant_hash, stat in stats['hash_to_stats'].items():
    xs[variant_hash] = {}
    ys[variant_hash] = {}
    # looping over every 10 epochs (frequency at which validation was performed)
    for t in list(zip(*[stat_seed.items() for stat_seed in stat])):
        # looping over the collected performance metrics
        for key in t[0][1].keys():
            if key not in xs[variant_hash]:
                xs[variant_hash][key] = []
            if key not in ys[variant_hash]:
                ys[variant_hash][key] = []
            avg_perf = sum(seed_t[1][key] for seed_t in t) / len(t)
            xs[variant_hash][key].append(t[0][0])
            ys[variant_hash][key].append(avg_perf)
```

```python
# plotting
for variant_hash, metrics in xs.items():
    if desired_variants and variant_hash not in desired_variants:
        continue
    for metric, x in metrics.items():
        if desired_metrics and metric not in desired_metrics:
            continue
        y = ys[variant_hash][metric]
        # plot y vs x with title that includes the hash and metric name
        plt.plot(x, y)
        plt.title('%s: %s' % (metric, variant_hash))
        print(variant_hash)
        plt.show()
```

```python
# single figure plotting
for variant_hash, metrics in xs.items():
    if desired_variants and variant_hash not in desired_variants:
        continue
    for metric, x in metrics.items():
        if desired_metrics and metric not in desired_metrics:
            continue
        y = ys[variant_hash][metric]
        # plot y vs x with title that includes the hash and metric name
        plt.plot(x, y, label='variant %s' % variant_hash)
        plt.title('%s: %s' % (metric, variant_hash))
        print(variant_hash)
plt.legend()
plt.show()
```
